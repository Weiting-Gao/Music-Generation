## Relative measurement: generalizes the result among features with various dimensions


```python
metrics_list={'pitch_count':[pc_1,pc_2,pc_3],'note_count':[nc_1,nc_2,nc_3],'total_pitch_class_histogram':[pch_1,pch_2,pch_3]
         ,'pitch_range':[pr_1,pr_2,pr_3],'pitch_interval':[pi_1,pi_2,pi_3],
         'avg_IOI':[ioi_1,ioi_2,ioi_3],'note_length_hist':[nlh_1,nlh_2,nlh_3],'note_length_transition_matrix':[nltm_1,nltm_2,nltm_3]}
keys=list(metrics_list)
```

```python
loo = LeaveOneOut()
loo.get_n_splits(np.arange(num_samples))
set1_intra = np.zeros((num_samples, len(metrics_list), num_samples-1))
set2_intra = np.zeros((num_samples, len(metrics_list), num_samples-1))
set3_intra = np.zeros((num_samples, len(metrics_list), num_samples-1))
for key,val in metrics_list.items():
    idx=keys.index(key)
    for train_index, test_index in loo.split(np.arange(num_samples)):
        set1_intra[test_index[0]][idx] = utils.c_dist(np.array(val[0])[test_index], np.array(val[0])[train_index])
        set2_intra[test_index[0]][idx] = utils.c_dist(np.array(val[1])[test_index], np.array(val[1])[train_index])
        set3_intra[test_index[0]][idx] = utils.c_dist(np.array(val[2])[test_index], np.array(val[2])[train_index])
```

```python
loo = LeaveOneOut()
loo.get_n_splits(np.arange(num_samples))
sets_inter1 = np.zeros((num_samples, len(metrics_list), num_samples))
sets_inter2 = np.zeros((num_samples, len(metrics_list), num_samples))
for key,val in metrics_list.items():
    idx=keys.index(key)
    for train_index, test_index in loo.split(np.arange(num_samples)):
        sets_inter1[test_index[0]][idx] = utils.c_dist(np.array(val[0])[test_index], np.array(val[1]))
        sets_inter2[test_index[0]][idx] = utils.c_dist(np.array(val[0])[test_index], np.array(val[2]))
```

visualization of intra-set and inter-set distances

```python
plot_set1_intra = np.transpose(set1_intra,(1, 0, 2)).reshape(len(metrics_list), -1)
plot_set2_intra = np.transpose(set2_intra,(1, 0, 2)).reshape(len(metrics_list), -1)
plot_set3_intra = np.transpose(set3_intra,(1, 0, 2)).reshape(len(metrics_list), -1)
plot_sets_inter1 = np.transpose(sets_inter1,(1, 0, 2)).reshape(len(metrics_list), -1)
plot_sets_inter2 = np.transpose(sets_inter2,(1, 0, 2)).reshape(len(metrics_list), -1)
for i in range(0,len(metrics_list)):
    sns.distplot(plot_set1_intra[i], kde_kws={"label":'intra Train'})
    sns.distplot(plot_sets_inter1[i], kde_kws={"label":'inter Train&Wavenet'})
    sns.distplot(plot_sets_inter2[i],kde_kws={"label":'inter Train&LSTM'})
   

    plt.title(list(metrics_list.keys())[i])
    plt.ylabel('Density')
    plt.xlabel('Euclidean distance')
    plt.show()
```

the difference of intra-set and inter-set distances.

```python
for i in range(0, len(metrics_list)):
    kld_1=utils.kl_dist(plot_set1_intra[i], plot_sets_inter1[i])
    oa_1=utils.overlap_area(plot_set1_intra[i], plot_sets_inter1[i])
    kld_2=utils.kl_dist(plot_set1_intra[i], plot_sets_inter2[i])
    oa_2=utils.overlap_area(plot_set1_intra[i], plot_sets_inter2[i])
    print(keys[i] + ':')
    print('------------------------')
    print(' Wavenet')
    print('  KullbackLeibler divergence:',kld_1)
    print('  Overlap area:', oa_1)
    print(' ratio:', oa_1/kld_1)
    
    print(' LSTM')
    print('  KullbackLeibler divergence:',kld_2)
    print('  Overlap area:', oa_2)
    print(' ratio:', oa_2/kld_2)
```
